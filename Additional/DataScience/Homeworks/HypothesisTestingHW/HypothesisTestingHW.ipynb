```python
% matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
from scipy.stats import ttest_ind
```

```python
heights_data = pd.read_table("heights.csv")
heights_data.columns = ["height"]
heights_data
```

```python
f, (ax1, ax2) = plt.subplots(2,1,figsize = (5,6))
ax1.hist(heights_data["height"], bins = 20)

ax2.boxplot(heights_data["height"])
plt.show()

print("Mean heigth = " + str(round(heights_data["height"].mean(),2)) + "cm")
print("Standart deviation = " + str(round(heights_data["height"].std(ddof = 1), 5)) + "cm")
```
